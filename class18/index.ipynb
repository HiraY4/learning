## Apply custom function on columns(Series Data)

```python
def grade (per:int) ->str:
  if per >= 90:
    return "A"
  elif per >= 80:
    return "B"
  elif per >= 70:
    return "C"
  elif per >= 60:
    return "D"
  elif per >= 50:
    return "E"
  else:
    return "F"
  
grade(70)
```

```python
df.head()
```

```python
df['percentage'].apply(grade)
# df['percentage'].sample()
```

```python
df['Grade'] = df['percentage'].apply(grade)
df
```
